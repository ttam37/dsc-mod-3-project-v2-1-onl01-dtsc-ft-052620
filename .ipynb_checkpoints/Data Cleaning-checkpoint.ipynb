```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

# Import Data

```python
# will only be using these 2 datasets which include the overview of each game and the team's information
# does not include player's information for each game
data_games = pd.read_csv('nba_games_data/games.csv')
data_teams = pd.read_csv('nba_games_data/teams.csv')
data_games_details = pd.read_csv('nba_games_data/games_details.csv')
```

# Add Additional Stats

```python
# create pivot table to aggregate the stats
additional_stats = pd.pivot_table(data_games_details, values=['STL','BLK','TO','PF'],
                       index=['GAME_ID','TEAM_ID'],aggfunc=np.sum).reset_index()
additional_stats.head()
```

```python
# merge dataframe with the additional stats (home)
df = data_games.merge(additional_stats[['GAME_ID','TEAM_ID','BLK','PF','STL','TO']], 
              left_on= ['GAME_ID','HOME_TEAM_ID'], 
              right_on = ['GAME_ID','TEAM_ID'])
df = df.rename(columns={'BLK': 'BLK_home','PF': 'PF_home','STL': 'STL_home','TO': 'TO_home'})
df.head()
```

```python
# merge dataframe with the additional stats (away)
df = df.merge(additional_stats[['GAME_ID','TEAM_ID','BLK','PF','STL','TO']], 
              left_on= ['GAME_ID','VISITOR_TEAM_ID'], 
              right_on = ['GAME_ID','TEAM_ID'])
df = df.rename(columns={'BLK': 'BLK_away','PF': 'PF_away','STL': 'STL_away','TO': 'TO_away'})
df.head()
```

```python
# remove overlapped columns
added_stats_df = df.drop(['TEAM_ID_x','TEAM_ID_y','TEAM_ID_x','TEAM_ID_y','TEAM_ID_home','TEAM_ID_away'],axis=1)
added_stats_df.head()
```

# Replace Team ID with Team Name

```python
# pd.pivot_table(a, values='TO',index=['GAME_ID','TEAM_ID'],aggfunc=np.sum).reset_index()
```

```python
added_stats_df.head()
```

```python
# replace home team_id with the team name for easier readability
final_data = added_stats_df.merge(data_teams[['TEAM_ID','NICKNAME']], left_on='HOME_TEAM_ID', right_on='TEAM_ID')
final_data = final_data.drop(['HOME_TEAM_ID'],axis=1)
final_data = final_data.rename(columns={"NICKNAME": "HOME_TEAM"})
final_data.head()
```

```python
# replace away team_id with the team name for easier readability
final_data = final_data.merge(data_teams[['TEAM_ID','NICKNAME']], left_on='VISITOR_TEAM_ID', right_on='TEAM_ID')
final_data = final_data.drop(['VISITOR_TEAM_ID','TEAM_ID_x','TEAM_ID_y'],axis=1)
final_data = final_data.rename(columns={"NICKNAME": "VISITOR_TEAM"})
final_data.head()
```

# 2. Data Cleaning

## Remove Duplicates

The only column that needs to be checked is GAME_ID to check if there are any duplicate entries.

```python
# print the number of duplicate entries
duplicates = final_data['GAME_ID'].duplicated().sum()
print(f'There are {duplicates} duplicate entries in GAME_ID.')
```

## Remove Features

```python
final_data.columns
```

```python
# removing unimportant features
final_data = final_data.drop(['GAME_ID','GAME_STATUS_TEXT','GAME_DATE_EST'],axis=1)
final_data.head()
```

```python
# # removing point features
# final_data = final_data.drop(['PTS_home','PTS_away'],axis=1)
# final_data.head()
```

## Missing Null Values

```python
def find_null_cols(df):
    df_clone = df.copy()
    print('\'NaN\' found in columns as follows:')
    for col in df_clone.columns:
        if (df_clone[col].isnull().any()) | (df_clone[col].isna().any()):
            print(f'\t{col:30} contains {df_clone[col].isna().sum():10} null values ({df_clone[col].isna().sum()*100/df_clone.shape[0]:5.2f}%) !')
    return None

print(final_data.shape)
find_null_cols(final_data)
```

```python
# the entire dataset has 99 rows with missing values which we will remove
final_data = final_data.dropna()

# check dataframe shape
print(final_data.shape)

# checking the null values were removed
find_null_cols(final_data)
```
